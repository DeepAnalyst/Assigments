```python
print("Enter the elements of matrix 1:")
m = int(input("Enter the number of rows: "))
n = int(input("Enter the number of columns: "))
matrix1 = []
for i in range(m):
    row = []
    for j in range(n):
        row.append(int(input("Enter element at position ({},{})".format(i+1, j+1))))
    matrix1.append(row)


print("Enter the elements of matrix 2:")
matrix2 = []
for i in range(m):
    row = []
    for j in range(n):
        row.append(int(input("Enter element at position ({},{})".format(i+1, j+1))))
    matrix2.append(row)

result = []
for i in range(m):
    row = []
    for j in range(n):
        row.append(matrix1[i][j] + matrix2[i][j])
    result.append(row)

print("The sum of the two matrices is:")
for row in result:
    print(row)
```

```python
print("Enter the elements of matrix 1:")
m1 = int(input("Enter the number of rows: "))
n1 = int(input("Enter the number of columns: "))
matrix1 = []
for i in range(m1):
    row = []
    for j in range(n1):
        row.append(int(input("Enter element at position ({},{})".format(i+1, j+1))))
    matrix1.append(row)


print("Enter the elements of matrix 2:")
m2 = int(input("Enter the number of rows: "))
n2 = int(input("Enter the number of columns: "))
matrix2 = []
for i in range(m2):
    row = []
    for j in range(n2):
        row.append(int(input("Enter element at position ({},{})".format(i+1, j+1))))
    matrix2.append(row)


if n1 != m2:
    print("The matrices cannot be multiplied.")
else:
  
    result = [[0 for j in range(n2)] for i in range(m1)]


    for i in range(m1):
        for j in range(n2):
            for k in range(m2):
                result[i][j] += matrix1[i][k] * matrix2[k][j]

    print("The product of the two matrices is:")
    for row in result:
        print(row)
```

```python
print("Enter the elements of the matrix:")
m = int(input("Enter the number of rows: "))
n = int(input("Enter the number of columns: "))
matrix = []
for i in range(m):
    row = []
    for j in range(n):
        row.append(int(input("Enter element at position ({},{})".format(i+1, j+1))))
    matrix.append(row)

transpose_matrix = [[matrix[j][i] for j in range(m)] for i in range(n)]

print("The original matrix is:")
for row in matrix:
    print(row)
print("The transpose of the matrix is:")
for row in transpose_matrix:
    print(row)
```

```python
string = input("Enter a string: ")

words = string.split()

sorted_words = sorted(words)

sorted_string = ' '.join(sorted_words)

print("The sorted string is:")
print(sorted_string)
```
